```python
import numpy as np
import pandas as pd
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data
```

```python
y = iris.target
```

```python
X.shape
```

```python
y.shape
```

```python
y
```

```python
from keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_trian, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler().fit(X_train)
```

```python
scalar
```

```python
X_scaled = scalar.transform(X_train)
```

```python
X_scaled
```

```python
X_test_scaled = scalar.transform(X_test)
```

```python
import keras
from keras.layers import Dense
from keras import Sequential
```

```python
model = Sequential()
```

```python
model.add(Dense(8,input_dim=4 ,activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_scaled, y_trian, epochs=150)
```

```python
predicted_classes = model.predict_classes(X_test_scaled)
```

```python
predicted_classes
```

```python
y_test.argmax(axis=1)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(classification_report(y_test.argmax(axis=1), predicted_classes))
```

```python
confusion_matrix(y_test.argmax(axis=1), predicted_classes)
```

```python
print(accuracy_score(y_test.argmax(axis=1), predicted_classes))
```

```python
model.save('iris_model_v1.h5')
```
